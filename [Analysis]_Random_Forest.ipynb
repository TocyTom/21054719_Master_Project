```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#Load the data 
hpLondon = pd.read_csv('London_HP_Final.csv')
hpLondon.head()
```

```python
#finding whther there is any missing values in the dataset.
hpLondon.info()
```

```python
hpLondon.isnull().sum()
```

```python
#finding whther there is any missing values in the dataset.
hpLondonNull = hpLondon.isnull().sum().sort_values(ascending=False)
hpLondonPercentage = (hpLondon.isnull().sum()/hpLondon.isnull().count()).sort_values(ascending = False)*100
hpLondonNull = pd.concat([hpLondonNull, hpLondonPercentage], axis = 1 , keys = ['Total number of values missing', 'Percentage of values missing'])
print('Below are the details of values missing in the dataframe')
hpLondonNull

```

```python
hpLondon['houses_sold'].fillna(hpLondon.groupby('area')['houses_sold'].transform('mean'), inplace = True) 

hpLondon['median_salary'].fillna(hpLondon.groupby('area')['median_salary'].transform('mean'), inplace = True) 


hpLondon['mean_salary'].fillna(hpLondon.groupby('area')['mean_salary'].transform('mean'), inplace = True) 


hpLondon['number_of_jobs'].fillna(hpLondon.groupby('area')['number_of_jobs'].transform('mean'), inplace = True) 


hpLondon['life_satisfaction'].fillna(hpLondon.groupby('area')['life_satisfaction'].transform('mean'), inplace = True) 

hpLondon['no_of_houses'].fillna(hpLondon.groupby('area')['no_of_houses'].transform('mean'), inplace = True) 

hpLondon['no_of_crimes'].fillna(hpLondon.groupby('area')['no_of_crimes'].transform('mean'), inplace = True) 

hpLondon['area_size'].fillna(hpLondon.groupby('area')['area_size'].transform('mean'), inplace = True) 

hpLondon['population_size'].fillna(hpLondon.groupby('area')['population_size'].transform('mean'), inplace = True) 


```
